```python
import pandas as pd
import numpy as np
```

```python
data_path = '../Data/IMDB_Dataset_Cleaned.csv'
df = pd.read_csv(data_path)
#df = df.rename(columns={'movie_title.1' : 'movie_title'})
#df = df.drop_duplicates(subset='movie_title').dropna()
df.head()
```

```python
df['genres'] = df['genres'].apply(lambda x: x.split(','))
```

```python
unique = ['Action', 'Adventure', 'Fantasy', 'Sci-Fi', 'Thriller', 'Romance', 'Animation', 'Comedy',
          'Family', 'Musical', 'Mystery', 'Western', 'Drama', 'History', 'Sport', 'Crime', 'Horror',
          'War', 'Biography', 'Music', 'Documentary', 'Short', 'Film-Noir' 
         ]

for col in unique:
    df[col] = 0
    
df.head()
```

```python
for idx in range(len(df)):
    genres = df.loc[idx, 'genres']
    for genre in genres:
        df.iloc[idx, genre] = 1

df = df.drop(['genres'], axis=1)
df[unique].head(20)
```

```python
df.to_csv('../Data/imdb_clean.csv')
```
